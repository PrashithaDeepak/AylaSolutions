```python
from pdf2image import convert_from_path
import pytesseract
import openpyxl
from openpyxl import Workbook
import re

# Specify the path to the Tesseract OCR executable
pytesseract.pytesseract.tesseract_cmd = r'/usr/bin/tesseract'  # Update the path as needed

# Convert PDF to images
pdf_path = '/file.pdf'
images = convert_from_path(pdf_path)

# Create a new workbook and select the active worksheet
workbook = Workbook()
sheet = workbook.active
sheet.title = 'Extracted Data'

# Initialize row number for writing in Excel
row_number = 1

# Loop through each image (page)
for i, image in enumerate(images):
    # Extract text from image
    text = pytesseract.image_to_string(image, lang='eng')

    # Write the text to Excel
    sheet.cell(row=row_number, column=1).value = f'Page {i + 1}'
    row_number += 1
    for line in text.split('\n'):
        # Remove illegal characters from the line
        line = re.sub(r'[\000-\010]|[\013-\014]|[\016-\037]', '', line)
        # Write the cleaned line to Excel
        sheet.cell(row=row_number, column=1).value = line
        row_number += 1

    # Leave a blank line between pages
    row_number += 1

# Save the Excel file
output_excel_path = '/extracted_text.xlsx'
workbook.save(output_excel_path)

print(f'Text extraction complete. Data saved to {output_excel_path}')
```

```python
from pdf2image import convert_from_path
import easyocr
import numpy as np
import cv2

def convert_pdf_to_img_and_extract_columns(pdf_file, num_columns=2, poppler_path='/usr/bin'):
    # Convert PDF to images (one image per page)
    images = convert_from_path(pdf_file, poppler_path=poppler_path)

    # Initialize EasyOCR reader
    reader = easyocr.Reader(['en'])

    all_text = []
    for page_number, img in enumerate(images):
        # Convert PIL Image to NumPy array for OpenCV processing
        img_np = np.array(img)
        height, width, _ = img_np.shape

        # Calculate the width of each column
        col_width = width // num_columns

        page_text = f"Page {page_number + 1}:\n"

        # Process each column and extract text
        for i in range(num_columns):
            # Crop the column
            col_img = img_np[:, i * col_width:(i + 1) * col_width]

            # Perform OCR on the column
            result = reader.readtext(col_img, detail=0)

            # Combine text for the column
            page_text += f"Column {i + 1}:\n" + "\n".join(result) + "\n\n"

        all_text.append(page_text)

    # Return all extracted text as a single string
    return "\n".join(all_text)

# Example usage:
pdf_text = convert_pdf_to_img_and_extract_columns('/content/drive/MyDrive/ayla/file.pdf')
print(pdf_text)
```

```python
from pdf2image import convert_from_path
import easyocr
import numpy as np

def convert_pdf_to_img_and_extract_columns(pdf_file, num_columns=2, poppler_path='/usr/bin'):
    # Convert PDF to images (one image per page)
    images = convert_from_path(pdf_file, poppler_path=poppler_path)

    # Initialize EasyOCR reader
    reader = easyocr.Reader(['en'])

    all_text = []
    for page_number, img in enumerate(images):
        # Convert PIL Image to NumPy array for OpenCV processing
        img_np = np.array(img)
        height, width, _ = img_np.shape

        # Define custom column widths for questions and answers based on the image's layout
        question_col_width = int(width * 0.40)  # 40% of the width for questions
        answer_col_start = question_col_width  # Start of the answer column

        page_text = f"Page {page_number + 1}:\n"

        # Extract questions (left column)
        question_col_img = img_np[:, :question_col_width]
        question_result = reader.readtext(question_col_img, detail=0)
        question_text = "Questions:\n" + "\n".join(question_result) + "\n\n"

        # Extract answers (right column)
        answer_col_img = img_np[:, answer_col_start:]
        answer_result = reader.readtext(answer_col_img, detail=0)
        answer_text = "Answers:\n" + "\n".join(answer_result) + "\n\n"

        # Combine question and answer text for the current page
        page_text += question_text + answer_text
        all_text.append(page_text)

    # Return all extracted text as a single string
    return "\n".join(all_text)

# Example usage:
pdf_text = convert_pdf_to_img_and_extract_columns('/content/drive/MyDrive/ayla/file.pdf')
print(pdf_text)
```
